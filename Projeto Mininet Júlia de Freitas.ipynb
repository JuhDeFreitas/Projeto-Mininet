#### Criação da Topologia da rede

Rede Linear com 6 Switches conectados em série, cada um com um host conectado à ele.

Cada host com um endereços MAC atribuido.

Links com largura de banda limitada a 25 Mbps usando o Traffic Control (TC).



- Elementos criados
 
  - Hosts :  h1, h2, h3, h4, h5 e h6

  - Switches : s1, s2, s3, s4, s5 e s6

  - Controlador : c0 (padrão mininet)


Comando para definição da topologia:


```python
sudo mn --topo linear,6 --mac --link tc,bw=25
```
